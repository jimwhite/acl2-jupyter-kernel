```python
; Test 1 - Arithmetic: should have symbols (+ referenced)
(+ 2 3)
```

```python
; Test 2 - Symbol fields: name, package, kind, operator, argument
(car '(a b c))
```

```python
; Test 3 - Operator position: CONS should be marked operator
(cons 'x '(y))
```

```python
; Test 4 - Defun with body symbols: should see CONSP, CONS, CAR in symbols
(defun exw-fn1 (x y) (if (consp x) (cons (car x) y) y))
```

```python
; Test 5 - Dependencies: exw-dep1 should depend on consp, car
(defun exw-dep1 (x) (if (consp x) (car x) x))
```

```python
; Test 6 - No dependencies for plain arithmetic
(+ 10 20)
```

```python
; Test 7 - Recursive function: deps should include consp, cdr (not self)
(defun exw-rec (x) (if (consp x) (exw-rec (cdr x)) 0))
```

```python
; Test 8 - Macro expansion: cw is a macro, should produce expansion
(cw "test~%")
```
